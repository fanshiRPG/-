```python
import requests as rq
from bs4 import BeautifulSoup as bs
```

```python
def getHTML(url):
    r=rq.get(url)
    r.encoding=r.apparent_encoding
    return r.text
import re
def getContent(html):
    soup=bs(html,'lxml')
    try:
        title=soup.find('h2')
        title=title.text
        time=soup.find('div',attrs={'align':'center'})
        time=time.text[3:time.text.find('来源')-1]
    except:
        title=''
        time=''
    try:
        content=soup.find('div',attrs={'id':re.compile('vsb_content')})
        content=content.text
    except:
        content=''
    return title,time,content

def getURL(html):
    soup=bs(html,'lxml')
    ls=soup.find_all('tr',attrs={'height':"20"})
    urls=[]
    for item in ls:
        urls.append('https://news.hbmu.edu.cn/'+item.a.get('href'))
    return urls
```

```python
article_urls=[]
article=[]
def main():
    page_url=['https://news.hbmu.edu.cn/xyyw.htm']
    for i in range(258,0,-1):
        page_url.append(f'https://news.hbmu.edu.cn/xyyw/{i}.htm')
    for page in page_url:
        page_html=getHTML(page)
        article_urls.extend(getURL(page_html))
    print(len(article_urls))
    i=0
    for url in article_urls:
        if i%10==0:
            print(i,'\t')
        i+=1
        article_html=getHTML(url)
        article.append(getContent(article_html))

main()
```
