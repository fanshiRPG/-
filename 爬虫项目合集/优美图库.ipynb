```python
import requests
from bs4 import BeautifulSoup
import re

url='https://www.umei.cc/'

resp=requests.get(url)
resp.encoding='utf-8'
a=BeautifulSoup(resp.text,'html.parser')
b=a.find('div',class_='sut_lbtC_qzdhc box_div').find_all('a')
for i in b:
    #print(i['href'])
    href=i.get('href')
    c=requests.get(url+href)
    c.encoding='utf-8'
    d=re.compile('<div class="picbox"><a href="(.*?)">.*?title=".*?"(.*?)</a></p>',re.S)
    e=re.findall(d,c.text)
    for j in e:
        print(url+str(j[0]),j[1])
```

```python
from bs4 import BeautifulSoup as bfs
import requests

for j in range(7):
    j=j+1
    url=f'http://www.umeituku.com/bizhitupian/weimeibizhi/{str(j)}.htm'
    resp=requests.get(url)
    resp.encoding='utf-8'
    b=resp.text
    a=bfs(b,'html.parser')
    c=a.find('div',class_='TypeList').find_all('img')
    for i in c:
        e=i.get('src')
        img_resp=requests.get(e)   #重新请求
        img_name=e.split('/')[-1]  #该文件 名字
        with open('1234/'+img_name,'wb') as f:  #1234‘/’表明文件路径的 ‘wb’二进制写入
            f.write(img_resp.content)  #这个方法就是。。。。。
```
