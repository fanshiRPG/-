```python
import requests
import time
import json

f=open('岗位.txt','w',encoding='utf-8')
url='https://careers.tencent.com/tencentcareer/api/post/Query?'
for j in range(1,43):
    params={
        'timestamp':int(time.time()*1000),
        'keyword':'python',
        'pageIndex': j,
        'pageSize': 10,
        'language':'zh-cn',
        'area':'cn',
        'countryId':'',
        'cityId':'',
        'bgIds':'',
        'productId':'',
        'categoryId':'',
        'parentCategoryId':'',
        'attrId':''
    }
    resp=requests.get(url,params)
    p=resp.json()
    items=p['Data']['Posts']
    for i in items:
        print(i)
        f.write(json.dumps(i))
f.close()
resp.close()
```

# 讲解json方法中的json.loads和json.dumps

```python
'''
首先了解一下json格式与python格式的异同：
json中只存在两种格式：
    1、对象，即字典形式；
    2、数组，即列表形式；
    其次是，python中保留字是True，而json中保留的是true
    并且，json中使用的是""来进行标识元素，而python‘’或者”“都可以进行标识
    
json.loads:将json数据转换成为python中的数据格式
json.dumps：将python数据转换程json数据格式

'''
with open('','a',encoding='utf-8') as f:
    j=f.read()
    print(j)
    d=json.loads(j)
    print(d)
# with open('','a',encoding='utf-8') as f:
#     j=json.dumps(d,ensure_ascii=False,indent=4)
#     f.write(j)
#     此处的indent表示的是使写入文本的数据进行4空格换位（缩进）/可以改变数字

```
